**Pandas DataFrames**

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

**Named Indexes**

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
```

# **Pandas - Plotting**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Automobile.csv')

df.plot()

plt.show()
```

**Scatter Plot**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Automobile.csv')

df.plot(kind = 'scatter', x = 'acceleration', y = 'horsepower')

plt.show()
```
